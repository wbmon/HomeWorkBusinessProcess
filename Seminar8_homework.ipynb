**Недавно бизнес-аналитик устроился на руководящую должность в компании AA по организации туров на велосипедах. Девять месяцев назад компания начала предоставлять новую услугу: сотрудники начали использовать электронную почту для уведомления клиентов о новых турах и предлагали скачать сведения о текущих планах туров и подробные инструкции.**

**Сначала клиенты были довольны новой услугой. Однако в последнее время многие начали жаловаться на получение устаревшей туристической информации. Аналитик понимает, что проблема заключается в процессе обновления информации и редактирования списка рассылки. Но он не знал, как решить эту проблему.**


### Задание 1 .
**Что бы сделали вы?**

### Решение.
Для начала попытался бы понять автоматизирован процесс рассылок или нет. И далее два пути.  
Если процесс автоматизирован, значит проблемы с формированием писем для рассылки связаны с неверно настроенными параметрами сбора информации для рассылки. здесь нужно разбираться с тем, как собирается информация о туристических предляжениях, как формируется список рассылки. Далее нужно смотреть кто осуществляет координацию этих бизнес-процессов и работать по их наладке. 
  
  
Если процесс осуществляется в ручном режиме, тогда важно понимать кто является маркетологом и формирует контент, задает дедлайн? Кто является копирайтером и пишет тексты для рассылок? Далее разбираемся кому попадают готовые тексты, т.е. кто дизайнер, который подбирает изображения, создает инфографику, а затем отдает готовый «сверстанный текст» маркетологу? Хотя, может быть, в компании АА нет отдельного дизайнера и эту функцию выполняет копирайтер. Или штатный маркетолог может быть вообще три в одном: и тексты пишет, и графику готовит, и стратегию продвижения продумывает. Здесь также нужно понять используются ли при формировании рассылки SEO-ключи, используется ли оптимизация. Таким образом, анализируя данные бизнес-процессы, можно пять где находятся "узкие места" и с чем оня связаны: с недостатьяной квалификацией, или с исполнительской дисциплиной.  
  
  
Здесь важно донести до ответственных лиц, что маркетинг не является какой-то разовой активностью, вроде «Один раз целевую аудиторию проанализировал и готово». В маркетинге важна постоянная работа и системность. Именно поэтому так важны поставленные бизнес-процессы. Только они могут гарантировать неизменное качество материалов и регулярность рассылок.

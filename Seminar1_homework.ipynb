### Задание 1 .
**Пропишите простыми словами ключевые отличия бизнес-модели от операционной.**

### Решение.
Бизнес-модель любой компании представляет собой обобщенное стратегическое описание деятельности компании, определяющее процессы функционирования бизнеса в настоящий момент и на перспективу. Это, можно сказать, глобальное представление механизмов работы бизнеса для достижения стратегических целей компании.  
Из этих механизмов можно выделять структурные, финансовые, операционные и другие. Таким образом операционная модель, по-моему мнению, является частью общей бизнес-модели организации.  
По предназначению, уровню детализации и контексту использования операционные модели также могут представлять собой более обобщенные (глобальные, стратегические, отраслевые и т.д.) и более детализированные (локальные, целевые, тактические).
Интерес и внимание к операционным моделям со стороны IT-специалистов также связан со стремлением разобраться в повседневном функционировании организации и предложить варианты по её совершенствованию с иcпользованием современных средств автоматизации и анализа.  
Таким образом, можно сказать, что ключевые отличия бизнес-модели от операционной заключаются в их направленности и масштабе.

### Задание 2.
**Перечислите самые важные, на ваш взгляд, элементы операционной модели и аргументируйте свой выбор.**

### Решение.

Самыми важными, на мой взгляд, элементами ОМ являются элементы "Персонал" и "Непрерывное совершенствование".

**Персонал.** От наличия и качества персонала зависит вцелом возможность функционирования бизнеса. Этот элемент включает в себя такие аспекты, как привлечение талантов, обучение и развитие, управление эффективностью и вовлечение сотрудников.

**Непрерывное совершенствование.** Использование совокупности механизмов постоянного совершенствования компании, систематическое
обновление всей операционной модели с целью повышения эффективности позволят любому бизнесу следовать современным трендам развития, избегать вляиния негативных внешних и внутренних факторов на развитие бизнеса.
